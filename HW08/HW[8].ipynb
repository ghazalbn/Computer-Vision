```python
# Import required packages
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
from PIL import Image

##################### your code ################################
import matplotlib.pyplot as plt

img = cv2.imread("royan.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.title("royan image")
plt.axis("off")
plt.show()
```

```python
r, threshold =  cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(threshold, cmap='gray')
plt.title('otsu')
plt.axis("off")
plt.show()
```

```python
structural_element = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
print(structural_element)
```

```python
dilation = cv2.dilate(threshold, structural_element, 6)
plt.imshow(dilation, cmap='gray')
plt.title("otsu + dilation")
plt.axis("off")
plt.show()
```

```python
print('original image:')
image_path_in_colab='royan.jpg'
text = pytesseract.image_to_string(image_path_in_colab, lang='fas')
print(text)
```

```python
print('otsu:')
text_otsu = pytesseract.image_to_string(dilation , lang='fas')
print(text_otsu)
```

```python
adapt_threshold =  cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 25, 3)
cv2.imwrite('../adapt__threshold.png', adapt_threshold)
plt.imshow(adapt_threshold, cmap='gray')
plt.title('adaptive threshold')
plt.axis("off")
plt.show()
```

```python
result = cv2.dilate(adapt_threshold, structural_element, 10)
cv2.imwrite('../result.png', adapt_threshold)
plt.imshow(result, cmap='gray')
plt.title("adaptive threshold + dilation")
plt.axis("off")
plt.show()
```

```python
################################################################
```

```python
print('adaptive thresholding:')
print(pytesseract.image_to_string( result , lang='fas'))
```

```python
# Import required packages
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
from PIL import Image
##################### your code ################################
img = cv2.imread("mypic.png")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.title("royan image")
plt.axis("off")
plt.show()
```

```python
r, threshold =  cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(threshold, cmap='gray')
plt.title('otsu')
plt.axis("off")
plt.show()
```

```python
structural_element = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
print(structural_element)
```

```python
dilation = cv2.dilate(threshold, structural_element, 6)
plt.imshow(dilation, cmap='gray')
plt.title("otsu + dilation")
plt.axis("off")
plt.show()
```

```python
adapt_threshold =  cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21, 2)
plt.imshow(adapt_threshold, cmap='gray')
plt.title('adaptive threshold')
plt.axis("off")
plt.show()
```

```python
result = cv2.dilate(adapt_threshold, structural_element, 6)
plt.imshow(result, cmap='gray')
plt.title("adaptive threshold + dilation")
plt.axis("off")
plt.show()
```

```python
################################################################
# Finding contours
im2 = img.copy()
contours, hierarchy = cv2.findContours(dilation, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)

print('otsu:')
for cnt in contours:
  x, y, w, h = cv2.boundingRect(cnt)
	# Drawing a rectangle on copied image
  rect = cv2.rectangle(im2, (x, y), (x + w, y + h), (0, 255, 0), 2)
	# Cropping the text block for giving input to OCR
  cropped = im2[y:y + h, x:x + w]
  print(pytesseract.image_to_string(cropped, lang='fas'))
```

```python
# Finding contours
im3 = img.copy()
contours, hierarchy = cv2.findContours(result, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)

print('adaptive thresholding:')
for cnt in contours:
  x, y, w, h = cv2.boundingRect(cnt)
	# Drawing a rectangle on copied image
  rect = cv2.rectangle(im3, (x, y), (x + w, y + h), (0, 255, 0), 2)
	# Cropping the text block for giving input to OCR
  cropped = im3[y:y + h, x:x + w]
  print(pytesseract.image_to_string(cropped, lang='fas'))
```
