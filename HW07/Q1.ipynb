# Panorama with OpenCV

```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
```

```python
# read input victoria images and show them in a row together
images = []
plt.figure(figsize=(30,4))
for i in range(1, 8):
    images.append(cv2.imread(f'images/victoria{i}.png'))
    plt.subplot(1, 8, i)
    plt.imshow(images[i - 1])
    plt.xticks([])
    plt.yticks([])
    plt.title(f'victoria{i}')
```

```python
# initialize OpenCV's image sticher object and then perform the image stitching on input images
stitcher = cv2.Stitcher_create(0)
status, stitched = stitcher.stitch(images)
```

```python
# show victoria panorama

# if the status is '0', then OpenCV successfully performed image
# stitching
if status == 0:
    plt.imshow(stitched)
    plt.axis('off')
    plt.title('victoria panorama')
    
# otherwise the stitching failed, likely due to not enough keypoints being detected
else:
    print("[INFO] image stitching failed ({})".format(status))
```
