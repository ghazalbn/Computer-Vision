```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros((32, 32))
image[31 , 3] = 255
plt.imshow(image, cmap='gray', vmax=255)
plt.axis('off')
```

```python
f = np.fft.fft2(image)
fshift = np.fft.fftshift(f)      # for change loc of f(0,0) to center

plt.subplot(1, 2, 1)
plt.imshow(np.real(fshift), cmap='gray')
plt.title('fft shifted Real')

plt.subplot(1, 2, 2)
plt.imshow(np.imag(fshift), cmap='gray')
plt.title('fft shifted Image')
plt.show()

magnitude_spectrum = 20 * np.log(np.abs(fshift))
plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('fftshift Magnitude Spectrum')
plt.show()
```
