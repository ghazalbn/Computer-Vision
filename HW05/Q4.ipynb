```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/img_02.jpg', 0 )
plt.imshow(image, cmap='gray')
print(np.unique(image))
```

```python
# for y in range(a):
#     for x in range(b):
#         if image[y, x] != 255:
#             points.append((y, x))

a, b = image.shape
points = [(y, x) for x in range(b) for y in range(a) if image[y, x] != 255]
print(image.shape)
print(len(points))
```

```python
ys = [p[0] for p in points]
xs = [p[1] for p in points]
y_m = np.mean(ys)
x_m = np.mean(xs)
xy_m = np.mean(np.multiply(ys, xs))
x_m_2 = x_m ** 2
x_2_m = np.mean(np.power(xs, 2))

m = np.float64((y_m * x_m - xy_m) / (x_m_2 - x_2_m))
c = y_m - m * x_m
print(f'y = {m}x + {c}')
```

```python
x = np.linspace(0, b)
data = cv2.imread('images/img_02.jpg')
y = m * x + c
plt.plot(x, y, '-r', label=f'y={m}x+{c}', linewidth=1)
plt.title(f'line on img_02.jpg')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.imshow(data)
plt.show()
```
