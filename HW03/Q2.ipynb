```python
# Import Your Libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## Part 1

### Your Histogram Equalization Function

```python
def hist_equ(image):
    '''
    input:
    image (ndarray): input image
    output:
    output_image (ndarray): enhanced image
    '''
    
    ############
    # Your code
    # Start
    
    output_image = image.copy()
    L = 256
    
    # Step 1
    histogram = np.zeros(L, np.int32)
    element, histogram[element] = np.unique(image, return_counts=True)
    
    # Step 2
    for i in range (1, L):
        histogram[i] += histogram[i-1]
        
    # Step 3
    n = image.shape[0] * image.shape[1]
    S = ((L - 1) / n) * histogram
    
    # Step 4
    S = np.round(S)
    
    # Step 5
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            output_image[i][j] = S[output_image[i][j]] 
            
    # End
    
    return output_image
```

```python
img = cv2.imread('River.jpg', 0)

### YOUR CODE ###
# START
equ = hist_equ(img)
# END

res = np.hstack((img, equ)) #stacking images side-by-side

plt.figure(figsize=(16, 16))
plt.imshow(res, cmap='gray')
```

## Part 2

### CLAHE OpenCV Library

```python
def CLAHE(image, gridSize, clip_limit):
    '''
    inputs:
    image (ndarray): input image
    gridSize (tuple): window size
    clip_limit (int): threshold for contrast limiting
    output:
    output_image (ndarray): improved image
    '''

    ############
    # Your code
    # Start
    
    clh = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=gridSize)
    output_image = clh.apply(image)
    
    # End

    return output_image
```

```python
img = cv2.imread('River.jpg', 0)

### YOUR CODE ###
# START
clh = CLAHE(img, (8, 8), 3)
# END

res = np.hstack((img, clh)) #stacking images side-by-side

plt.figure(figsize=(16, 16))
plt.imshow(res, cmap='gray')
```
